```python
import pandas as pd
import numpy as np #importing the required library
```

```python
dataset = pd.read_csv("DigitalAd_dataset.csv")
dataset.head(100) #load the dataset
```

```python
print(dataset.shape)
print(dataset.head(10)) # verify the dataset
```

```python
x = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values # splitting the dataset columns into x,y columns. x contains 2 columns, y contains 1 column as a output column
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.20,random_state=126)#train and test for the split dataset
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
print(x_train)#this is for the standardized the column values into a single based column value
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train,y_train) # train the model logistic regression
```

```python
y_pred = model.predict(x_test)
#using x_test the y_pred should be predict
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)

print("accuracy score of the model: {0}%".format(accuracy_score(y_test, y_pred)*100))#verify the accuracy of the model
```

```python
age = int(input("Enter the customer age :"))
sal = int(input("Enter the customer salary :"))
newcust = [[age,sal]]
result = model.predict(sc.transform(newcust))
print(result)
if result == 1:
    print("The customer will buy")
else:
    print("The customer will not buy")# give the input values and check wheather this is working or not.
```
